```python
import numpy as np

import pandas as pd

from sklearn.preprocessing import MinMaxScaler

from sklearn.linear_model import  Ridge, LinearRegression, Lasso

from sklearn.model_selection import train_test_split

from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

```python
# import dataset

df = pd.read_csv(r'C:\Users\MICHEAL\Desktop\HAMOYE_DS_INTERNSHIPS\energydata_complete.csv')

# view first five rows
df.head(5)
```

```python
df.info()
```

```python
# drop unimportant columns

df = df.drop(['date', 'lights'], axis=1)
df.head(5)
```

```python
# normalise data

scaler = MinMaxScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(df.values), columns=df.columns)
scaled_df.head(5)
```

```python
# target variable

pred_target = scaled_df['Appliances']
scaled_df = scaled_df.drop(columns=['Appliances'],axis=1)


# splitting data into train and test

X_train, X_test, y_train, y_test = train_test_split(scaled_df, pred_target, random_state=42, test_size=0.3)
scaled_df.head(5)
```

```python
# Question 12
from sklearn.metrics import r2_score
linear_reg = LinearRegression()
linear_reg.fit(X_train[['T2']], X_train.T6)
pred_y = linear_reg.predict(X_test[['T2']])
r2_score = r2_score(X_test.T6, pred_y)
round(r2_score,2)
```

```python
from sklearn.metrics import mean_absolute_error
linear_reg.fit(X_train,y_train)
pred_target = linear_reg.predict(X_test)
mae = mean_absolute_error(y_test,pred_target)
round(mae,2)
```

```python
linear_reg.fit(X_train, y_train)
pred_target = linear_reg.predict(X_test)
rss = np.sum(np.square(y_test - pred_target))
round(rss,2)
```

```python
from sklearn.metrics import mean_squared_error
linear_reg.fit(X_train, y_train)
pred_target = linear_reg.predict(X_test)
mse = mean_squared_error(y_test, pred_target)
round(mse,3)
```

```python
# weights of a model

def get_weights_df(model, features, col_name) :
    weights = pd.Series(model.coef_, features.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = [ 'Features' , col_name]
    weights_df[col_name].round( 3 )
    return weights_df
```

```python
linear_model_weights = get_weights_df(linear_reg, X_train, 'Linear_Model_Weight' )
linear_model_weights.sort_values(by=['Linear_Model_Weight'])
```
